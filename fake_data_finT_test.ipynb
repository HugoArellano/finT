```python
import pandas as pd
import numpy as np
from faker.providers.person.en import Provider
```

```python
df = pd.DataFrame({'usuario': [], 'nombre': [], 'dependientes':[] ,'propiedad_casa': [], 'propiedad_auto': [],
                  'habilidad_futuro':[], 'habilidad_manejo':[], 'habilidad_conceptos':[],
                  'estres_dinero':[], 'estres_cotidiano':[],'deuda_instru':[], 'deuda_gen':[], 'deuda_dti':[],
                  'ingresos':[], 'ahorro_disciplina':[], 'ahorro_desempleo':[],
                  'inversion_fin':[], 'inversion_nofin':[], 'contacto':[], 'medio_finT':[] })
```

```python
df['usuario'] = np.linspace(1001,1021,20).astype(int)
```

```python
def random_names(name_type, size):
    names = getattr(Provider, name_type)
    return np.random.choice(names, size=size)

df['nombre'] = random_names('first_names',20)
```

```python
from numpy.random import default_rng
rng = default_rng()
```

```python
df['dependientes'] = rng.integers(low=0, high=4, size=20)
```

```python
df['propiedad_casa'] = rng.integers(low=1, high=4, size=20)
df['propiedad_auto'] = rng.integers(low=1, high=4, size=20)
```

```python
df['habilidad_futuro'] = rng.integers(low=1, high=4, size=20)
df['habilidad_manejo'] = rng.integers(low=1, high=6, size=20)
df['habilidad_conceptos'] =  rng.integers(low=1, high=4, size=20)
```

```python
df['estres_dinero'] = rng.integers(low=1, high=4, size=20)
df['estres_cotidiano'] = rng.integers(low=1, high=6, size=20)
```

```python
df['deuda_instru'] = rng.integers(low=1, high=5, size=20)
df['deuda_gen'] = rng.integers(low=1, high=4, size=20)
df['deuda_dti'] = rng.integers(low=1, high=5, size=20)
```

```python
df['ingresos'] = rng.integers(low=1, high=4, size=20)
```

```python
df['ahorro_disciplina'] = rng.integers(low=1, high=5, size=20)
df['ahorro_desempleo'] = rng.integers(low=1, high=4, size=20)
```
